```python
n = 7
max_weight = 15
input_items = [(10, 2), (5, 3), (15, 5), (7, 7), (6, 1), (18, 4), (3, 1)]
population_size = 4
mutation_rate = 0.8
epoch = 200
verbose = 1
```

```python
import random

class Item:
    def __init__(self, profit, weight):
        self.profit = profit
        self.weight = weight

def get_items(n, input_items=None, verbos=0):
    items = []
    
    if input_items is None:
        for i in range(n):
            print(f"Item {i+1}")
            item_profit = int(input("Profit item: "))
            item_weight = int(input("Weight item: "))
            items.append(Item(item_profit, item_weight))
    else:
        for item in input_items:
            items.append(Item(item[0], item[1]))

    if verbos:
        for idx, item in enumerate(items):
            print(f'Item {idx + 1}: Profit {item.profit}, Weight {item.weight}')
            
    return items

def init_population(n, population_size):
    population_list = []
    
    for i in range(population_size):
        new_individual = [random.choice([0, 1]) for _ in range(n)]
        new_individual = new_individual + [None, None]
        population_list.append(new_individual)
        
    return population_list

def cross_over(population_list, n, p):
    n_b2 = n // 2
    for i in range(0, p, 2):
        child1 = population_list[i][:n_b2] + population_list[i+1][n_b2:n] + [None, None]
        child2 = population_list[i+1][:n_b2] + population_list[i][n_b2:n] + [None, None]
        population_list.append(child1)
        population_list.append(child2)
    return population_list

def mutation(population_list, n, p, m):
    chosen_ones = [i for i in range(p, p * 2)]
    random.shuffle(chosen_ones)
    chosen_ones = chosen_ones[:int(((p * 2) - 1) * m)]
    for i in chosen_ones:
        cell = random.randint(0, n - 1)
        population_list[i][cell] = 1 if population_list[i][cell] == 0 else 0
        
    return population_list

def weight(bag, n, max_weight, item_list):
    total_weight = 0
    for i in range(n):
        if bag[i]:
            total_weight += item_list[i].weight
    return abs(max_weight - total_weight) if total_weight > max_weight else 200

def profit(bag, n, item_list):
    total_profit = 0
    for i in range(n):
        if bag[i]:
            total_profit += item_list[i].profit
    return total_profit

def fitness(population_list, n, p, item_list, max_weight):
    for i in range(p * 2):
        population_list[i][n] = weight(population_list[i], n, max_weight, item_list)
        population_list[i][n+1] = profit(population_list[i], n, item_list)
    
    return population_list


def sorter(population_list,index1,index2):
    soreted_list = sorted(population_list,key = lambda x: (x[index1],-x[index2]))
    
    return soreted_list



if __name__ == "__main__":
    n = 7
    max_weight = 15
    input_items = [(10, 2), (5, 3), (15, 5), (7, 7), (6, 1), (18, 4), (3, 1)]
    population_size = 4
    mutation_rate = 0.8
    epoch = 200
    verbose = 1  # Verbose mode

    print(f'Max weight is {max_weight}')
    items = get_items(n, input_items=input_items, verbos=verbose)
    
    while epoch:

        current_population = init_population(n, population_size)
        current_population = cross_over(current_population, n, population_size)
        current_population = mutation(current_population, n, population_size, mutation_rate)
        current_population = fitness(current_population, n, population_size, items, max_weight)
        current_population = sorter(current_population,n,n+1)
        current_population = current_population[:population_size]
        epoch-=1
        for i in current_population:
            print(i)
        break
    else:
        print('best found solotion:',current_population[0])
```
